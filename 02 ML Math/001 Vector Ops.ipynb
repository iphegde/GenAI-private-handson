Basics - Vector Operations in Python

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
#Ploting a Vector

plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1 )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1 , color='g' )
plt.quiver(0,0,-4,-5, scale_units='xy', angles='xy', scale=1 , color='y' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
print( vector_1 + vector_2 )
```

Dot product of 2 vectors is a Scalar

```python
a = np.array([2,3])

b = np.array([4,4])

a_dot_b = np.dot( a, b )

print(a_dot_b)
```

Cross product of 2 vectors is a Vector

```python
a = np.array([2,3,10])

b = np.array([4,4,12])

a_cross_b = np.cross(a , b)

print(a_cross_b)
```

Projection of 'a' vector  on 'v' vector  = Resultant is a VECTOR
 

```python
a = np.array([2,5])

v = np.array([8,-6])

mag_of_v = np.sqrt(sum(v**2))

# Above formula is coded like below
proj_of_a_on_v = (np.dot(a,v)/mag_of_v**2)*v

print(proj_of_a_on_v)
```
